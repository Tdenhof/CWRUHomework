```python


def formatData(df):
    df = df.reset_index()
    final = []
    for i in range(len(df)):
        lst = df.loc[i]
        temp = []
        for i in lst:
            temp.append(str(i))
        temp.pop(0)
        string = ' '.join(temp)
        final.append(string)
    return final 

def writeData(df):
    for i in range(1,11):
        trainName = 'train' + str(i) + '.txt'
        testName = 'test' + str(i) + '.txt'
        df = df.sample(frac = 1)
        train = df.iloc[:11250,:]
        test = df.iloc[11250:,:]
        train = formatData(train)
        test = formatData(test)
        dfTrain = pd.DataFrame(train)
        dfTest = pd.DataFrame(test)
        tfile = open(trainName,'w')
        tfile.write(dfTrain.to_string(index_names= False, index = False))
        tstfile = open(testName,'w')
        tstfile.write(dfTest.to_string(index_names= False, index = False))
        tstfile.close()

def adaboostDataTran(df):
    for i in range(1,65):
        new = []
        tempList = df[i]
        for x in range(len(df[i])):
            values = tempList[x].split(':')
            new.append(values[1])
        df[i] = new
    return df
        

```
